```python
# Librerias
import pandas as pd
import numpy as np
```

```python
# Ruta al archivo
ruta = "../data/raw/201911_EnvigmuBDD_casadas_unidas.csv"  

# Lectura del CSV
df = pd.read_csv(
    ruta,
    encoding="cp1252",
    low_memory=False
)
```

```python
# Vision preliminar del DF
df.head()
```

```python
# Aislamos Columnas referentes al analisis

columnas_objetivo = [
    # Niñez
    "f2_s5_1",
    "f2_s5_3",
    "f2_s5_5",

    # Violencia de pareja actual
    "f2_s7a_14_16",
    "f2_s7a_14_17",
    "f2_s7a_14_18",
    "f2_s7a_14_19",
    "f2_s7a_14_20"
]

# Nota: Esto conllevo un previo analisis del total de 1144 preguntas
```

```python
# Revisamos si todas las col. existen dentro del DF
column_presentes = [column for column in columnas_objetivo if column in df.columns]
column_faltantes = [column for column in columnas_objetivo if column not in df.columns]
print("Presentes:", column_presentes)
print("Faltantes:", column_faltantes)
```

```python
# Revisamos el DF con las col. que necesitamos
DF_firt_grade = df[column_presentes]
DF_firt_grade.head()
```

```python
# Informacion general sobre el DF
DF_firt_grade.describe()
```

```python
DF_firt_grade.info()
```

```python
# Normalizar los datos
DF_firt_grade = DF_firt_grade.apply(lambda col: col.astype(str).str.strip().str.lower())
# Nota: Eliminamos espacios en blanco, convertimos en string y minusculas
```

```python
# Identificamos cuantos valores existen en cada col. para cada respuesta
for col in DF_firt_grade:
    print("\n")
    print(DF_firt_grade[col].value_counts(dropna=False))
```

```python
# En porcentajes a cuanto equivale cada respuesta
for col in DF_firt_grade:
    print("\n")
    print(DF_firt_grade[col].value_counts(normalize = True, dropna = False ).mul(100).round(2))
```

- El grado de respuestas 'no recuerdo' es menor al < 3% lo cual es muy bajo y no compromete al analisis. Pero no lo descartamos 
- Las observaciones con valores faltantes o ‘no aplica’ fueron excluidas debido a su incidencia marginal (<0.1%).

```python
# Cambiamos el estado "no aplica" por null
DF_second_grade = DF_firt_grade[~DF_firt_grade.isin(['no aplica'])]
```

```python
# Cambiamos el string 'na' por un NULL de pandas y eliminarlo
DF_second_grade = DF_second_grade.replace("nan", np.nan)
```

```python
# Eliminamos los valores Null
DF_second_grade = DF_second_grade.dropna()
```

```python
len(DF_second_grade)
```

```python
# Confirmamos la eliminamos de valores no necesarios
for col in DF_second_grade:
    print("\n")
    print(DF_second_grade[col].value_counts(normalize = True, dropna = False ).mul(100).round(2))
```

```python
# Exortamos el DF limpio 
DF_second_grade.to_csv(
    "../data/processed/df_limpio_violencia.csv",
    index = False,
    encoding="cp1252"
)
```
