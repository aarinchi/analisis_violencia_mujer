# Contexto del estudio:
- La violencia contra la mujer es un problema social complejo que puede manifestarse en distintas etapas de la vida y en diversos ámbitos. Comprender los factores asociados a su persistencia es clave para el diseño de políticas públicas y para el análisis social basado en evidencia.
- En este proyecto se analiza información proveniente de la Encuesta Nacional sobre Relaciones Familiares y Violencia de Género contra las Mujeres, realizada por el INEC (Ecuador) en el año 2019.
- El análisis se enfoca específicamente en el Formulario 2 (F2), aplicado a mujeres casadas o unidas, el cual es el más extenso del levantamiento y contiene información detallada sobre experiencias de violencia a lo largo del ciclo de vida.

# Fuente de Datos:
- Institución: Instituto Nacional de Estadística y Censos (INEC)
- Año: 2019
- Cobertura geográfica: Costa, Sierra y Oriente
- Población objetivo: Mujeres casadas o unidas
- Registros totales: 10.590
- Número total de preguntas (F2): 1.144

Fuente Oficial (link):
https://www.ecuadorencifras.gob.ec/violencia-de-genero/

# Alcance del análisis
Dado el gran alcance de las variables disponibles a analizar hemos seleccionado un subconjunto específico de preguntas apermiten analizar la relación entre:
- Exposición a violencia durante la infancia, dentro del entorno familiar
- Experiencias de violencia en la vida de pareja durante la adultez

Para este fin, se seleccionaron 7 preguntas clave relacionadas con:
- Violencia en la familia de origen
- Violencia física con su pareja actual.

Las variables que no aportaban directamente al objetivo del análisis fueron desestimadas, priorizando claridad y relevancia analítica.

# Pregunta de investigación
La pregunta central de este proyecto es:

¿Existe una relación entre haber vivido violencia en la infancia y experimentar violencia con la pareja actual durante la adultez en las mujeres?

Esta pregunta se aborda con un enfoque observacional y descriptivo, sin asumir relaciones causales directas.

# Objetivo del Análisis 
## Objetivo General
- Analizar si existe una asociación entre la exposición a violencia en la infancia y la probabilidad de experimentar violencia en la pareja durante la adultez, utilizando datos oficiales del INEC.

## Objetivos Específicos
- Comprender la estructura y características del dataset seleccionado
- Limpiar y preparar las variables relevantes para el análisis
- Definir criterios claros para la clasificación de violencia
- Aplicar análisis estadístico descriptivo para evaluar la relación entre ambas etapas de la vida
- Interpretar los resultados de forma responsable, considerando las limitaciones del estudio


# Enfoque metodológico
El análisis se desarrollará a partir de las siguientes etapas:

- Exploración y limpieza de datos
- Definición de variables analíticas binarias
- Construcción de tablas de contingencia
- Cálculo de probabilidades y Odds Ratio
- Visualización e interpretación de resultados

# Consideraciones éticas y analíticas

Este proyecto tiene un enfoque estrictamente analítico.
Los resultados obtenidos:

- No pretenden generalizarse fuera del contexto del estudio
- No establecen causalidad
- No sustituyen análisis cualitativos 

El objetivo es demostrar cómo el análisis de datos puede aportar evidencia cuantitativa para comprender de mejor manera fenómenos sociales complejos.
